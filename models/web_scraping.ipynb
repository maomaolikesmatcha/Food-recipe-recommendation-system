```python
# Import the required libraries 
import pandas as pd
from bs4 import BeautifulSoup 
import requests
import time 
```

```python
# Define the url in python 
url = 'https://www.epicurious.com/search?content=recipe'
```

```python
# Fetching html from the website
page = requests.get(url)
# BeautifulSoup enables to find the elements/tags in a webpage 
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup
```

```python
links = []
for link in soup.find_all('a', class_='view-complete-item'):
    links.append(link.get('href'))
```
